Masukkan Library yang digunakan

```python
import pandas as pd
import numpy as np
```

Memanggil Dataset

```python
data = pd.read_csv('results_cleaned.csv')
data.head()
```

Deskripsi Dataset

```python
data.info()
```

Seleksi Fitur

```python
features = ['building_area (m2)', 'land_area', 'bedroom_count', 'bathroom_count', 'carport_count']
x = data[features]
y = data['price']
x.shape, y.shape
```

Split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=70)
x_test.shape,y_test.shape
```

Membuat Model Regresi Linear

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test, y_test)
print('akurasi model regresi linear', score)
```

Membuat Input Model Regresi Linear

```python
input_data = np.array([[220, 220, 3, 3, 0]])

prediction = lr.predict(input_data)
print('estimasi harga rumah dalam Rp :', prediction)
```

Saving Model

```python
import pickle

filename = 'estimasi_rumah.sav'
pickle.dump(lr,open(filename,'wb'))
```
